```python
tasks = {
    "Monday": [],
    "Tuesday": [],
    "Wednesday": [],
    "Thursday": [],
    "Friday": [],
    "Saturday": [],
    "Sunday": []
}
```

```python
while True:
    print("\n--- Weekly Planner ---")
    print("1. Add task")
    print("2. View tasks")
    print("3. Delete task")
    print("4. Edit task")
    print("5. Exit")

    choice = input("Choose: ")

    # ➤ Add task
    if choice == "1":
        print("Add a task")
        day = input("Enter day (Monday-Sunday): ").capitalize()
        task = input("Enter task: ")

        if day in tasks:
            tasks[day].append(task)
            print("Task added!")
        else:
            print("Invalid day!")

    # ➤ View tasks
    if choice == "2":
        print("\nYour Tasks:")
        for day in tasks:
            if tasks[day]:
                print(day + ":")
                for t in tasks[day]:
                    print(" -", t)

    # ➤ Delete task
    if choice == "3":
        print("Delete a task")
        day = input("Enter day: ").capitalize()

        if day in tasks:
            if tasks[day]:
                for i, t in enumerate(tasks[day]):
                    print(i, "-", t)

                index = int(input("Enter task number to delete: "))
                tasks[day].pop(index)

                print("Task deleted!")
            else:
                print("No tasks for this day!")
        else:
            print("Invalid day!")

    # ➤ Edit task
    if choice == "4":
        print("Edit a task")
        day = input("Enter day: ").capitalize()

        if day in tasks:
            if tasks[day]:
                for i, t in enumerate(tasks[day]):
                    print(i, "-", t)

                index = int(input("Enter task number to edit: "))
                new_task = input("Enter new task: ")

                tasks[day][index] = new_task
                print("Task updated!")
            else:
                print("No tasks for this day!")
        else:
            print("Invalid day!")

    # ➤ Exit
    if choice == "5":
        print("Goodbye!")
        break
```
